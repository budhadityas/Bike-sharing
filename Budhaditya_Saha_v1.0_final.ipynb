# Step 4: Training the model (approach: forward sequencing)

I will be using statsmodels to fit a regression line through the training data

```python
import statsmodels.api as sm

# Add a constant
X_train_lm = sm.add_constant(X_train['atemp'])

# Create a first fitted model
lr = sm.OLS(y_train, X_train_lm).fit()
```

```python
# check the parameters obtained

lr.params
```

```python
# visualize the data with the scatter plot and the fitted regression line

plt.scatter(X_train_lm.iloc[:, 1], y_train)
plt.plot(X_train_lm.iloc[:, 1], 0.163 + 0.683*X_train_lm.iloc[:, 1], 'r')
plt.show()
```

```python
lr.summary()
```

Observations:
1. The p-value is 0 which implies that the model is statistically significant.
2. The R-squared is 0.418 which signifies that 41.8 % of the variance in "cnt" could be explained by the predictor "atemp" which is good.

### Adding another variable

The next predictor variable having the highest co-efficient ratio with cnt is yr

```python
X_train_lm = X_train[['atemp', 'yr']]

# Add a constant
X_train_lm = sm.add_constant(X_train_lm)

# Create a first fitted model
lr = sm.OLS(y_train, X_train_lm).fit()

# Check the summary
lr.summary()
```

Observations:
1.Adding the new variable "yr" still has all p-values as 0.000 which implies the model is statistically significant
2.The new variable has a positive co-efficient
3.The R-Squared value has increased to 69.6 which implies that 69.6% of the variance in count is explained by the 2 predictor variables atemp & year which is preety good.

### Adding another variable

The next predictor variable having the highest co-efficient ratio with cnt is Aug

```python
X_train_lm = X_train[['atemp', 'yr', 'Aug']]

# Add a constant
X_train_lm = sm.add_constant(X_train_lm)

# Create a first fitted model
lr = sm.OLS(y_train, X_train_lm).fit()

# Check the summary
lr.summary()
```

3.Hence the new variable is not a significant predictor
